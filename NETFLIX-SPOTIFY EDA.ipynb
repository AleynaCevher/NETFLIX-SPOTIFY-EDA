```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

List the top 10 most listened-to artists on Spotify.

```python
df = pd.DataFrame(data)
df_streams = df.groupby('Artist')['Stream'].sum().sort_values(ascending=False)[:10]
fig, ax2 = plt.subplots(figsize=(8, 6))
ax2.set_title('Top 10 most listened')
df_streams.plot(kind='bar', color='maroon', ax=ax2)
ax2.set_xlabel('Artist')
ax2.set_ylabel('Streams')
plt.tight_layout()
plt.show()
```

List the top 10 most listened-to artists on Spotify.

```python
df = pd.DataFrame(data)
df_views = df.groupby('Artist')['Views'].sum().sort_values(ascending=False)[:10]
fig, ax1 = plt.subplots(figsize=(8, 6))
ax1.set_title('Top 10 most listened')
df_views.plot(kind='bar',color='green', ax=ax1)
ax1.set_xlabel('Artist')
ax1.set_ylabel('Views')
plt.tight_layout()
plt.show()
```

List the top 10 most-watched songs on Youtube.

```python
youtube = df.sort_values(by = 'Views', ascending = False)[['Track', 'Artist', 'Views']].head(12)
plt.figure(figsize=(27,9))
sns.barplot(data = youtube, x= 'Track', y='Views', palette = 'Greens_r')
plt.title("Top 10 most-watched songs")
plt.ylabel("Stream", color = 'lightgrey')
plt.xlabel("Song", color = 'lightgrey')
plt.show()
```

List the top 10 most-streamed songs on Spotify.

```python

spotify = df.sort_values(by = 'Stream', ascending = False)[['Track', 'Artist', 'Stream']].head(12)
plt.figure(figsize=(32,9))
sns.barplot(data = spotify, x= 'Track', y='Stream', palette = 'Reds_r')
plt.title("En Çok Dinlenen 10 Şarkı")
plt.ylabel("İzlenme", color = 'lightgrey')
plt.xlabel("Şarkı", color = 'lightgrey')
plt.show()

```

Show the percentage distribution of album genres.

```python
colors = ['#ff6f61', '#6b5b95', '#88b04b']

labels = ['Albüm', 'Single', 'EP']
sizes = [72.45, 23.90, 3.64]

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=80, colors=colors)
plt.title('Distribution of Album Genres')
plt.legend(labels, loc='best')
plt.axis('equal')
plt.show()
```

Show the distribution of characteristics scores.

```python
numeric_cols = data.select_dtypes(include='number')
numeric_cols.hist(bins=50,color='orange', figsize=(20,15))
```

Show the distribution of characteristics in the most-watched songs on YouTube.

```python

Youtube_En_Çok_8 = df.sort_values('Views', ascending=False).reset_index().head(8)
melt_youtube = Youtube_En_Çok_8.melt(id_vars='Track', value_vars=['Danceability', 'Energy', 'Speechiness', 'Acousticness', 'Valence'], var_name='characteristic', value_name='value')
plt.figure(figsize=(28, 13))
sns.pointplot(x='Track', y='value', hue='characteristic', data=melt_youtube, dodge=True, markers=["o", "s", "D", "^", "v"], linestyles=["-", "--", "-.", ":", "-"])
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1))
plt.title('The Characteristic Features of the 8 Most-Watched Songs on YouTube')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Show the distribution of characteristics in the most-watched songs on YouTube.

```python
Spotify_En_Çok_8 = df.sort_values('Stream', ascending=False).reset_index().head(10)
melt_spotify = Spotify_En_Çok_8.melt(id_vars='Track', value_vars=['Danceability', 'Energy', 'Speechiness', 'Acousticness', 'Valence'], var_name='characteristic', value_name='value')

plt.figure(figsize=(28, 13))
sns.pointplot(x='Track', y='value', hue='characteristic', data=melt_spotify, dodge=True, markers=["o", "s", "D", "^", "v"], linestyles=["-", "--", "-.", ":", "-"])
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1))
plt.title('The Characteristic Features of the 8 Most-Watched Songs on YouTube')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
